```python
import speech_recognition as sr
from gtts import gTTS
from IPython.display import Audio, display
import io
import time

def speak(text):
    tts = gTTS(text=text)
    fp = io.BytesIO()
    tts.write_to_fp(fp)
    fp.seek(0)
    audio_data = fp.read()
    display(Audio(audio_data, autoplay=True))
    time.sleep(len(text.split()) * 0.6 + 1.5)

def listen():
    recognizer = sr.Recognizer()
    recognizer.energy_threshold = 300
    recognizer.pause_threshold = 1.0
    recognizer.dynamic_energy_threshold = True

    try:
        with sr.Microphone() as source:
            print("Adjusting for background noise...")
            recognizer.adjust_for_ambient_noise(source, duration=2)
            print("You can speak now...")
            time.sleep(0.8)  # small delay before recording
            audio = recognizer.listen(source, timeout=10, phrase_time_limit=8)
    except sr.WaitTimeoutError:
        print("Timeout. No voice detected.")
        return ""
    except Exception as e:
        print("Microphone issue:", e)
        return ""

    try:
        response = recognizer.recognize_google(audio)
        print("Recognized:", response)
        return response.lower()
    except sr.UnknownValueError:
        print("Sorry, I did not understand.")
        return ""
    except sr.RequestError:
        print("Speech Recognition service error.")
        return ""

def listen_with_retry(retries=3):
    for attempt in range(retries):
        text = listen()
        if text:
            return text
        print("I could not catch that. Please try again.")
        speak("I could not understand. Please say it again.")
    return ""

questions = [
    {"question": "What is the capital of India?", "answer": "new delhi"},
    {"question": "What is two plus two?", "answer": "4"},
    {"question": "Which planet is known as the red planet?", "answer": "mars"},
    {"question": "What is the color of the sky?", "answer": "blue"},
    {"question": "What is the opposite of hot?", "answer": "cold"}
]

score = 0
speak("Welcome to the Smart Audio Quiz.")
print("Welcome to the Smart Audio Quiz.")

speak("What is your name?")
name = listen_with_retry().title()
if not name:
    name = "Student"

speak(f"Hello {name}, we are starting your quiz now.")
print(f"Hello {name}, we are starting your quiz now.")

for index, item in enumerate(questions):
    time.sleep(1)
    speak(f"Question {index + 1}: {item['question']}")
    print(f"Question {index + 1}: {item['question']}")
    speak("Now please answer.")
    time.sleep(1)  # Give extra time before listening

    user_answer = listen_with_retry()

    if item["answer"].lower() in user_answer:
        speak("Correct.")
        print("Correct.")
        score += 1
    else:
        speak(f"Incorrect. The correct answer is {item['answer']}.")
        print(f"Incorrect. The correct answer is: {item['answer']}")

    time.sleep(1.5)

speak(f"{name}, your quiz is complete. You scored {score} out of {len(questions)}.")
print(f"{name}, you scored {score} out of {len(questions)}.")
speak("Thank you for playing the Smart Audio Quiz.")
print("Thank you for playing the Smart Audio Quiz.")
```
